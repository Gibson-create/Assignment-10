```python
#Importing the libraries 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
#loading the data
df = pd.read_csv("KAG_conversion_data.csv")
df.head()
```

```python
df.info()
```

```python
#converting the categorical variables to  dummy variables
X = pd.get_dummies(df)
X.head()
```

```python
X.info()
```

```python

X.drop(['ad_id', 'xyz_campaign_id' , 'fb_campaign_id', 'interest'], axis=1)
X.head ()
```

```python
#Using a heat map to show the correlation 

X.corr()
```

```python
#heat map to show correlation
plt.figure(figsize=(15, 5))
heatmap = sns.heatmap(X.corr(), vmin=-1, vmax=1,annot=True )
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':11}, pad=12);
```

The columns were dropped because the corellation values the conversion variables are closer to zero showing that their is a non-linear relation between the variables and the target variables

```python
#Determine the best number of clusters for the data in the range of 2 to 20 using the elbow method
from sklearn.cluster import KMeans
wcss = []
for i in range(2, 20):
    km = KMeans(n_clusters = i, init = 'k-means++', max_iter = 500, n_init = 10, random_state = 0)
    km.fit(X)
    wcss.append(km.inertia_)
    import warnings
warnings.filterwarnings('ignore')
```

```python
#graphical plot for the elbow curve
plt.plot(range(2, 20), wcss)
plt.xlabel('No. of Clusters')
plt.ylabel('wcss')
plt.title('No of clusters for the data (range of 2 to 20)', fontsize = 12)
```

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster.elbow import kelbow_visualizer

# Use the quick method and immediately show the figure
kelbow_visualizer(KMeans(random_state=4), X, k=(2,20))
```

```python
# k = 5
km = KMeans(n_clusters=5)
```

```python
km.fit(X)
```

```python
# Fit n Predict
clusters = km.predict(X)
```

```python
print (clusters)
```

```python
# Getting all the Labels
y_means= km.labels_
```

```python
y_means
```

```python
#Adding the values in y_means to the original data frame (not X) as column ‘Advert_Type’

df['Advert_Type'] = y_means.tolist()
df.head()
```

```python
#Which advert type lead to the highest and consistent amount of sales by customers of all the age brackets? 
sns.barplot(y='Approved_Conversion',x='Advert_Type',data=df)
```
